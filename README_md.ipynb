This project uses **Groq** (free tier, no credit card):

3. Click **API Keys** → **Create API Key**

```
User Query
    │
    ├──→ Embed (HuggingFace, local) → ChromaDB → Top-K Chunks
    │                                                  │
    └──────────────────────────────────────────────────┤
                                                       ▼
                                              RAG Prompt Template
                                                       │
                                                       ▼
                                              LLM — Groq Llama 3.3
                                                       │
                                                       ▼
                                                 Final Answer ✅
```
